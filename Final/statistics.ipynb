```python
train = pd.read_csv("./data/train.csv")

train.head()
```

```python
print(train['ConstructName'].nunique())
print(train['SubjectName'].nunique())
```

```python
id_columns = ['MisconceptionAId', 'MisconceptionBId', 'MisconceptionCId', 'MisconceptionDId']

empty_counts = {}
for col in id_columns:
    empty_count = train[col].isna().sum()
    empty_counts[col] = empty_count

print("各欄位空值數量：")
for col, count in empty_counts.items():
    print(f"{col}: {count} 個空值")

total_empty = sum(empty_counts.values())
print(f"\n四個欄位總共有 {total_empty} 個空值")
```

```python
id_columns = ['MisconceptionAId', 'MisconceptionBId', 'MisconceptionCId', 'MisconceptionDId']

id_counts = {}
for col in id_columns:
    col_counts = train[col].value_counts()
    
    for id_val, count in col_counts.items():
        if pd.notna(id_val):
            if id_val not in id_counts:
                id_counts[id_val] = count
            else:
                id_counts[id_val] += count

max_occurrences = max(id_counts.values()) if id_counts else 0

id_categories = {
    f'出現{i}次': [] for i in range(1, max_occurrences + 1)
}
id_categories['從未出現'] = set(range(2569)) - set(id_counts.keys())

for id_val, count in id_counts.items():
    id_categories[f'出現{count}次'].append(id_val)

for category, ids in id_categories.items():
    print(f"{category}: {len(ids)} 個 ID")
```
